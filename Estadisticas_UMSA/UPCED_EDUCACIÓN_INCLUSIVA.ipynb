##Estudiantes con discapacidad que ingresaron a la UMSA

```python
import matplotlib.pyplot as plt
import math  # Importamos el módulo math para usar cos y sin

# Datos de la gráfica
etiquetas = [
    "AÑO 2012", "AÑO 2013", "AÑO 2014", "AÑO 2015",
    "AÑO 2016", "AÑO 2017", "AÑO 2018", "AÑO 2019",
    "AÑO 2020", "AÑO 2021"
]
valores = [12, 17, 22, 20, 22, 19, 28, 45, 19, 16]
colores = [
    "#ffcccc", "#cce7ff", "#ffffcc", "#ffd9b3",
    "#d9b38c", "#ff9999", "#b3d9ff", "#b3c6ff",
    "#b3b3ff", "#dab3ff"
]  # Colores más claros para visibilidad
explode = [0 for _ in etiquetas]  # Ningún segmento resaltado

# Crear la gráfica
plt.figure(figsize=(8, 6))  # Tamaño ajustado para Colab
wedges, texts, autotexts = plt.pie(
    valores, labels=etiquetas, autopct='%1.0f%%', startangle=90,
    colors=colores, explode=explode, shadow=True, pctdistance=0.8
)

# Ajustar los valores exactos de ingresos (fuera del gráfico)
for i, wedge in enumerate(wedges):
    # Calcular el ángulo medio del segmento
    angle = (wedge.theta2 + wedge.theta1) / 2
    # Calcular coordenadas (x, y) para posicionar las etiquetas
    x = math.cos(math.radians(angle)) * 1.2  # Convertir ángulo a radianes
    y = math.sin(math.radians(angle)) * 1.2
    plt.text(x, y, str(valores[i]), horizontalalignment='center', fontsize=10)

# Mejorar la visibilidad de etiquetas dentro de la gráfica
for text in autotexts:
    text.set_color("black")
    text.set_fontsize(10)

# Título de la gráfica
plt.title("Distribución por Año con Número de Ingresos", fontsize=16)

# Mostrar la gráfica
plt.tight_layout()  # Ajustar diseño para que no se recorten elementos
plt.show()
```

##Datos sociodemográficos de los estudiantes que ingresaron a la UMSA gestiones  2012 - 2021.

```python
import matplotlib.pyplot as plt
import math  # Para los cálculos trigonométricos

# Datos de la gráfica
etiquetas = ["MASCULINO", "FEMENINO"]
valores = [132, 88]  # Cantidades
colores = ["#4A90E2", "#F5A623"]  # Azul y naranja
explode = [0, 0]  # Sin segmentos resaltados

# Crear la gráfica
plt.figure(figsize=(8, 6))  # Ajuste de tamaño
wedges, texts, autotexts = plt.pie(
    valores, labels=etiquetas, autopct='%1.0f%%', startangle=90,
    colors=colores, explode=explode, shadow=True, pctdistance=0.8
)

# Colocar valores exactos fuera del gráfico
for i, wedge in enumerate(wedges):
    angle = (wedge.theta2 + wedge.theta1) / 2  # Ángulo medio
    x = math.cos(math.radians(angle)) * 1.2  # Coordenadas x
    y = math.sin(math.radians(angle)) * 1.2  # Coordenadas y
    plt.text(x, y, str(valores[i]), horizontalalignment='center', fontsize=10)

# Ajustar el texto dentro del gráfico
for text in autotexts:
    text.set_color("black")
    text.set_fontsize(10)

# Título de la gráfica
plt.title("Gráfico N° 2: Género de los estudiantes", fontsize=16)

# Mostrar la gráfica
plt.tight_layout()
plt.show()
```

##Edades de las personas con discapacidad que ingresaron a la UMSA

```python
import matplotlib.pyplot as plt
import numpy as np

# Datos del gráfico
rangos = [
    "17 a 21 A", "22 a 26 A", "27 a 31 A", "32 a 36 A", "37 a 41 A",
    "42 a 46 A", "47 a 51 A", "52 a 56 A", "57 a 61 A", "62 a 66 A"
]
valores = [25, 26, 19, 17, 8, 2, 1, 1, 1, 1]  # Porcentajes
colores = ["#4A90E2", "#F5A623", "#7ED321", "#F8E71C", "#D0021B",
           "#8B572A", "#BD10E0", "#50E3C2", "#B8E986", "#4A4A4A"]

# Configurar la figura
plt.figure(figsize=(10, 6))

# Crear la gráfica de barras horizontales
y_pos = np.arange(len(rangos))
plt.barh(y_pos, valores, color=colores, edgecolor="black")

# Añadir etiquetas
for i in range(len(valores)):
    plt.text(valores[i] + 1, y_pos[i], f"{valores[i]}%", va='center', fontsize=9)

# Personalización
plt.yticks(y_pos, rangos, fontsize=10)
plt.xlabel("Porcentaje (%)", fontsize=12)
plt.title("Gráfico N° 3: Rango de edades", fontsize=16)
plt.grid(axis='x', linestyle='--', alpha=0.7)

# Mostrar la gráfica
plt.tight_layout()
plt.show()
```

##Tipos de discapacidad de las personas que ingresaron a la UMSA

```python
import matplotlib.pyplot as plt
import numpy as np

# Datos
categorias = [
    "Auditiva", "Baja Visión", "Ceguera Visual Total", "Ceguera Visual",
    "Física", "Física Motora", "Física Visceral", "Intelectual",
    "Mental Psíquica o Psicológica", "Múltiple Física Motora",
    "Múltiple Intelectual", "No Vidente", "Psicológica", "Sensorial"
]
valores = [14, 9, 16, 13, 1, 131, 2, 17, 3, 5, 1, 1, 3, 4]
porcentajes = [round((v / 220) * 100, 1) for v in valores]  # Calcular porcentajes

# Crear gráfica de barras horizontales
plt.figure(figsize=(12, 8))
y_pos = np.arange(len(categorias))
bars = plt.barh(y_pos, valores, color=plt.cm.Pastel1.colors)

# Añadir etiquetas con valores y porcentajes
for bar, valor, p in zip(bars, valores, porcentajes):
    width = bar.get_width()  # Longitud de la barra
    label_text = f"{valor} ({p}%)"  # Texto combinado
    plt.text(
        width + 2,  # Separar un poco del extremo de la barra
        bar.get_y() + bar.get_height() / 2,
        label_text,
        va="center", ha="left",
        color="black", fontsize=9
    )

# Configuración de la gráfica
plt.yticks(y_pos, categorias)
plt.xlabel("Número de Estudiantes")
plt.title("Distribución de Tipos de Discapacidad en la UMSA")
plt.gca().invert_yaxis()  # Invertir el eje Y para mejorar legibilidad
plt.tight_layout()

# Mostrar gráfica
plt.show()
```

##Ingresos por Facultad

```python
import pandas as pd
import matplotlib.pyplot as plt

# Datos
datos = {
    "Facultad": [
        "Facultad de Arquitectura y Artes",
        "Facultad de Ciencias Puras y Naturales",
        "Facultad de Ciencias Sociales",
        "Facultad de Ciencias Económicas y Financieras",
        "Facultad de Ciencias Farmacéuticas y Bioquímicas",
        "Facultad de Derecho y Ciencias Políticas",
        "Facultad de Humanidades y Ciencias de la Educación",
        "Facultad de Ingeniería",
        "Facultad de Medicina",
        "Facultad de Tecnología"
    ],
    "Total Estudiantes": [8, 18, 34, 32, 3, 44, 40, 4, 16, 6]
}

# Crear DataFrame
df = pd.DataFrame(datos)

# Calcular porcentajes
df["Porcentaje"] = (df["Total Estudiantes"] / df["Total Estudiantes"].sum()) * 100

# Crear gráfico de barras horizontales
plt.figure(figsize=(10, 8))
bars = plt.barh(df["Facultad"], df["Total Estudiantes"], color=plt.cm.Pastel1.colors)

# Añadir los porcentajes al lado de las barras
for i, bar in enumerate(bars):
    plt.text(
        bar.get_width() + 1,  # Colocar el texto al lado de la barra
        bar.get_y() + bar.get_height() / 2,
        f'{df["Porcentaje"][i]:.1f}%',  # Mostrar el porcentaje
        va='center', ha='left', fontsize=12
    )

# Título y etiquetas
plt.title("Distribución de Estudiantes por Facultad", fontsize=14)
plt.xlabel("Cantidad de Estudiantes", fontsize=12)
plt.ylabel("Facultad", fontsize=12)

plt.tight_layout()

# Mostrar gráfico
plt.show()
```

##Situación académica de los estudiantes con discapacidad en la UMSA

```python
import matplotlib.pyplot as plt

# Datos
estado_situacion = {
    "Estado": ["Abandono", "Continúan en Carrera", "Se Tituló", "Egreso"],
    "Cantidad": [117, 89, 10, 4]
}

# Crear gráfico de barras verticales
plt.figure(figsize=(8, 6))
bars = plt.bar(
    estado_situacion["Estado"],
    estado_situacion["Cantidad"],
    color=plt.cm.Pastel1.colors[:len(estado_situacion["Estado"])],  # Colores diferentes
    edgecolor='black'
)

# Añadir las cantidades y porcentajes encima de cada barra
total_estudiantes = sum(estado_situacion["Cantidad"])
for bar, cantidad in zip(bars, estado_situacion["Cantidad"]):
    percentage = (cantidad / total_estudiantes) * 100
    plt.text(
        bar.get_x() + bar.get_width() / 2,
        bar.get_height() + 2,
        f'{cantidad} ({percentage:.1f}%)',
        ha='center',
        va='bottom',
        fontsize=12
    )

# Título
plt.title(f"Estado de Situación Académica\nTotal Estudiantes: {total_estudiantes}", fontsize=14)

# Mejorar la estética
plt.xlabel("Estado Académico")
plt.ylabel("Cantidad de Estudiantes")
plt.xticks(rotation=45)
plt.tight_layout()

# Mostrar gráfico
plt.show()
```
